```python
import os
import re
from urllib import request
from urllib.error import HTTPError
from bs4 import BeautifulSoup
import requests


from moviepy.editor import VideoFileClip
from pydub import AudioSegment
```

```python
overlap=[]
url_f = 'https://www.fss.or.kr/fss/bbs/B0000203/list.do?menuNo=200686&bbsId=&cl1Cd=&pageIndex='  # 크롤링할 금감원 사이트 몸통 앞 부분
url_b = '&sdate=&edate=&searchCnd=1&searchWrd='  # 크롤링할 금감원 사이트 몸통 뒷 부분
site = 'https://www.fss.or.kr'   #가져올 사이트 앞부분
rec = "/fss/bbs/B0000203/view.do?" #반복되는 부분 뒷부분
dl = "/fss/cmmn/vodApi/apiVodDownload.do?"   #클릭해서 다운 받는 부분
```

```python
def Search(getA):
   for getLink in getA:
       data = getLink.get('href')
       #print(data)
       global count
       try:
           if rec in getLink.get("href"):
               if data not in overlap:
                   count += 1
                   overlap.append(data)
                   accessUrl = site + getLink.get("href")
                   r = requests.get(accessUrl)
                   soup = BeautifulSoup(r.text, "html.parser")
                   getDLATag = soup.find_all("a")
                   downSearch(getDLATag, count)


       except:pass
```

```python
def downSearch(getDLATag, count):
   for getDLLink in getDLATag:
       try:
           if dl in getDLLink.get('href'):
               print("\n다운로드 링크 : {}".format(site) + getDLLink.get('href'))
               accessDLUrl = site + getDLLink.get('href')
               #fileOriginalNM = re.sub('<.+?>', '', str(getDLLink), 0).strip().replace('_', ' ')
               fileNM = "금감원_인간음성_" + str(count)
               if os.path.isfile(file_path + fileNM):
                   print("다운로드 실패 : 동일 파일 존재\n")
               else:
                   get_download(accessDLUrl, fileNM, file_path)
       except:pass
```

```python
def mp4_to_wav(input_file, output_file):
    # 비디오 파일에서 오디오 추출
    video = VideoFileClip(input_file)
    audio = video.audio
    audio.write_audiofile("temp_audio.mp3")

    # 추출된 오디오 파일을 WAV로 변환
    sound = AudioSegment.from_mp3("temp_audio.mp3")
    sound.export(output_file, format="wav")

# 파일 경로 설정
#input_file = "path/to/your/input.mp4"
#output_file = "path/to/your/output.wav"

# 변환 실행
#mp4_to_wav(input_file, output_file)
```

```python
def get_download(url, fname, directory):
   try:
       os.chdir(directory)
       request.urlretrieve(url, fname)
       print('다운로드 완료')

       #다운로드한 mp4 파일을 wav 파일로 바꾼 후 mp4파일 삭제하기 (용량 이슈 + audio data만 필요하기 때문)
       outname = fname+".wav"
       mp4_to_wav(fname, outname)
       os.remove(fname)

   except HTTPError as e:
       print('error')

       return
```
